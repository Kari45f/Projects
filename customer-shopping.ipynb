🕵🏼‍♀️ **Exploratory Data Analysis (EDA)**

**Purchases by age group**

```python
df = dataset.groupby('Age')['Purchase Amount (USD)'].sum().reset_index()
```

```python
df
```

```python
plt.figure(figsize=(12,5))
sns.barplot(
    data=df,
    x='Age',
    y='Purchase Amount (USD)',
    palette='Set2')
plt.title('Purchases by Age Group')
plt.ylabel('Purchase Amount (USD)')
plt.tight_layout()
plt.show()
```

Purchasing trends indicate that 49-year-old have the highest spending habits, contrasting with 67-year-old group, who exhibits the lowest purchase volume. 
In summary, the groups with major purchases are the following:
* 49
* 69
* 25
* 41

Analysis reveals that individuals in their 49s have the highest purchasing capacity, in addition to the other groups indicated above.

🏘 **Groups with major purchasing power per location**

After the results obtained above, now, it's interesting to find out where these people come from.

```python
location = dataset.groupby('Location')['Purchase Amount (USD)'].sum().reset_index()
```

```python
location
```

```python
plt.figure(figsize=(12,5))
sns.barplot(
    data=location,
    x='Location',
    y='Purchase Amount (USD)',
    palette='Set2')
plt.title('Purchase amount per location')
plt.ylabel('Purchase Amount (USD)')
plt.xlabel('Location')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

The location with the highest score is Montana.

👫 **Who buys more, male or female?**

```python
top = dataset.groupby(['Item Purchased', 'Category', 'Gender']).size().reset_index(name='count')
```

```python
top.sort_values(by='count', ascending=False)
```

In the previous table, it's possible to see the gender that makes most purchases is "Male", but, this may happen due to the number of women & men registered in the dataset are not equal.

*Total people registered*

```python
total_people = dataset.groupby(['Gender']).size().reset_index(name='count')
```

```python
total_people
```

With this result,we can understand that the number of people registered is not equal, so we cannot conclude which gender buys more.

🔝 **Top product categories**

```python
top_category = dataset['Category'].value_counts()
colors= plt.cm.Set2(range(len(top_category)))
top_category.plot(kind='barh', color=colors)
plt.title("Top Product Categories")
plt.show()
```

Obtained results: People prefer to buy clothes rather than outerwear.

💸 **Most used payment method**

```python
pym_mtd = dataset['Payment Method'].value_counts().reset_index()
pym_mtd
```

```python
colors=plt.cm.Set2(range(len('pym_mtd')))

plt.figure(figsize=(8, 8))
plt.pie(
    pym_mtd['count'],
    labels=pym_mtd['Payment Method'],
    autopct='%1.1f%%',
    startangle=90,
    colors=colors
)
plt.title('Most use Payment Method')
plt.show()
```
